## 1. Checking if two given trees are identical??!!

** Two trees are identical when they have the same data and the arrangment of data is also the same.

```python
class Node:
    
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
def is_identical_trees(tree1, tree2):
    
    # Case 1: if trees are empty ---> return True
    if tree1 is None and tree2 is None:
        return True
    
    # Case 2: if trees are not empty ---> Compare them
    if tree1 is not None and tree2 is not None:
        return ((tree1.data == tree2.data) and
               is_identical_trees(tree1.left, tree2.left) and 
               is_identical_trees(tree1.right, tree2.right))
    
    # Case 3: if one tree is empty and another not
    return False
        
# Example 1: Two identical trees
tree1 = Node(1)
tree1.left = Node(2)
tree1.right = Node(3)
tree1.left.left = Node(4)
tree1.left.right = Node(5)

tree2 = Node(1)
tree2.left = Node(2)
tree2.right = Node(3)
tree2.left.left = Node(4)
tree2.left.right = Node(5)

# Example 2: Two empty trees
# tree1 = Node(None)
# tree2 = Node(None)

# Example 3: Not identical trees
# tree1 = Node(1)
# tree1.left = Node(2)

# tree2 = Node(2)
# tree2.right = Node(1)

# Example 4: Not identical trees (modeification of example 3)
# tree1 = Node(1)
# tree1.left = Node(2)

# tree2 = Node(1)
# tree2.right = Node(2)

if __name__ == '__main__':
    if is_identical_trees(tree1, tree2):
        print('Both trees are identical.')
    else:
        print('Not identical trees (-_-)!')
```

### Time complexity:

#### O(min(N, M))

N & M = sizes of the trees

## 2. Now let's check if two trees are identical or not by compairing their traversal??!!

To traverse the tree, there are tree options:

   - Preorder
   - Inorder
   - Postorder

If we just traverse by preorder, can we say two trees are identical? (YES/NO? Why?)

```python
class Node:
    
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
def preOrder_(tree, sol):
    if tree is None:
        return
    sol.append(tree.data)
    preOrder_(tree.left, sol)
    preOrder_(tree.right, sol)
    
def inOrder_(tree, sol):
    if tree is None:
        return
    inOrder_(tree.left, sol)
    sol.append(tree.data)
    inOrder_(tree.right, sol)

def postOrder_(tree, sol):
    if tree is None:
        return
    postOrder_(tree.left, sol)
    postOrder_(tree.right, sol)
    sol.append(tree.data)
    
def is_identical_trees(tree1, tree2):
    
    travInfo_tree1, travInfo_tree2 = [], []
    
    # Check the pre-order
    preOrder_(tree1, travInfo_tree1)
    preOrder_(tree2, travInfo_tree2)
    if travInfo_tree1 != travInfo_tree2:
        return False

    # Check the in-order
    travInfo_tree1.clear()
    travInfo_tree2.clear()
    
    inOrder_(tree1, travInfo_tree1)
    inOrder_(tree2, travInfo_tree2)
    if travInfo_tree1 != travInfo_tree2:
        return False
    
    # Check the in-order
    travInfo_tree1.clear()
    travInfo_tree2.clear()
    
    postOrder_(tree1, travInfo_tree1)
    postOrder_(tree2, travInfo_tree2)
    if travInfo_tree1 != travInfo_tree2:
        return False
    
    return True
    
    
# Example 1: Two identical trees
# tree1 = Node(1)
# tree1.left = Node(2)
# tree1.right = Node(3)
# tree1.left.left = Node(4)
# tree1.left.right = Node(5)

# tree2 = Node(1)
# tree2.left = Node(2)
# tree2.right = Node(3)
# tree2.left.left = Node(4)
# tree2.left.right = Node(5)

# Example 2: Two empty trees
# tree1 = Node(None)
# tree2 = Node(None)

# Example 3: Not identical trees
# tree1 = Node(1)
# tree1.left = Node(2)

# tree2 = Node(2)
# tree2.right = Node(1)

# Example 4: Not identical trees (modeification of example 3)
tree1 = Node(1)
tree1.left = Node(2)

if __name__ == '__main__':
    if is_identical_trees(tree1, tree2):
        print('Both trees are identical.')
    else:
        print('Not identical trees (-_-)!')
```

### Time complexity:

#### O(N)

## 3. Another way to determine if two trees are identical is just using preorder!! How??

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
def preOrder_(tree, sol):
    if tree is None:
        return
    sol.append(tree.data)
    
    if tree.left:
        preOrder_(tree.left, sol)
    else:
        sol.append('NO!')
    
    if tree.right:
        preOrder_(tree.right, sol)
    else:
        sol.append('NO!')
    
def is_identical_trees(tree1, tree2):
    
    travInfo_tree1, travInfo_tree2 = [], []
    
    # Check the pre-order
    preOrder_(tree1, travInfo_tree1)
    preOrder_(tree2, travInfo_tree2)
    return travInfo_tree1 == travInfo_tree2
    
    
# Example 1: Two identical trees
# tree1 = Node(1)
# tree1.left = Node(2)
# tree1.right = Node(3)
# tree1.left.left = Node(4)
# tree1.left.right = Node(5)

# tree2 = Node(1)
# tree2.left = Node(2)
# tree2.right = Node(3)
# tree2.left.left = Node(4)
# tree2.left.right = Node(5)

# Example 2: Two empty trees
# tree1 = Node(None)
# tree2 = Node(None)

# Example 3: Not identical trees
# tree1 = Node(1)
# tree1.left = Node(2)

# tree2 = Node(2)
# tree2.right = Node(1)

# Example 4: Not identical trees (modeification of example 3)
tree1 = Node(1)
tree1.left = Node(2)

tree2 = Node(1)
tree2.right = Node(2)

if __name__ == '__main__':
    if is_identical_trees(tree1, tree2):
        print('Both trees are identical.')
    else:
        print('Not identical trees (-_-)!')
```
